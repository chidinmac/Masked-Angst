```python
isDepressive = df['Indicator'] == "Symptoms of Depressive Disorder"
isAnxiety = df['Indicator'] == "Symptoms of Anxiety Disorder"
isBoth = df['Indicator'] == "Symptoms of Anxiety Disorder or Depressive Disorder"

depressiveDF = df[isDepressive] 
anxietyDF = df[isAnxiety]
bothDF = df[isBoth]
anxietyDF.head()

isByState = df["Group"] == "By State"
stateDF = df[isByState]

```

```python
isNE = depressiveDF['Group'] == 'National Estimate'
```

```python
NE = depressiveDF[isNE]
```

```python
values = us_state_abbrev.values()
stateList = list(values)
temp = {'state': {}, 'mean':[]}
means = []
temp['state'] = stateList
trimean = {'mean1': [], 'mean2':[] 'mean3'}

for i in range(3):
    for state in stateList:
        isState = (stateDF['State'] == state) & (stateDF['Phase'] == str(i))
        statedf = stateDF[isState]
        means.append(statedf['Value'].mean())

temp['mean'] = means

newDF = pd.DataFrame(temp)

# creating the chart object
horizontal_chart = pygal.HorizontalBar(spacing=20)
  
# naming the title
horizontal_chart.title = 'Horizontal Chart'        

horizontal_chart.add('', newDF['mean'])
horizontal_chart.x_labels = stateList

print(newDF.head())
horizontal_chart
stateDF.head()
```

```python
#Get Figures & Axis
fig, ax = plt.subplots()

#Plot Set up Chart
plt.plot(NE['Time Period Label'], NE['Value'], color='red', marker ='o')
plt.title('National Average of Depression Per Time Time Period', fontsize=14)
plt.xlabel('Time Period Label', fontsize=14)
plt.ylabel('Depression Average', fontsize=14)
plt.grid(True)
plt.rcParams["figure.figsize"] = (40,10)
plt.xticks(rotation = 45)

#Only display every nth x axis label 
for label in ax.xaxis.get_ticklabels()[1::2]:
    label.set_visible(False)
l = ax.xaxis.get_ticklabels()
l[0].set_visible(True)

#Show
plt.show()
```
